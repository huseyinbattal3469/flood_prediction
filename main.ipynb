```python
import warnings
from sklearn.exceptions import ConvergenceWarning, FitFailedWarning
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.naive_bayes import GaussianNB, BernoulliNB, CategoricalNB
from sklearn.linear_model import LogisticRegression, SGDClassifier, Ridge, Lasso, ElasticNet, SGDRegressor, LinearRegression
from sklearn.svm import SVC, LinearSVC, SVR, NuSVR, LinearSVR
from sklearn.neighbors import KNeighborsClassifier, RadiusNeighborsClassifier, KNeighborsRegressor
from sklearn.tree import DecisionTreeClassifier, ExtraTreeClassifier, DecisionTreeRegressor, ExtraTreeRegressor
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier, GradientBoostingClassifier, HistGradientBoostingClassifier, AdaBoostClassifier, RandomForestRegressor, BaggingRegressor, GradientBoostingRegressor, HistGradientBoostingRegressor, AdaBoostRegressor
from sklearn.neural_network import MLPClassifier, MLPRegressor
from sklearn.gaussian_process import GaussianProcessRegressor
from xgboost import XGBClassifier, XGBRegressor
from lightgbm import LGBMClassifier, LGBMRegressor
from catboost import CatBoostClassifier, CatBoostRegressor

def base_models(X, y, cv=5, scoring=["neg_mean_absolute_error", "neg_mean_squared_error", "neg_root_mean_squared_error", "r2", "explained_variance"], is_classifier=False,
                random_state=40, n_jobs=1):
    """
    :param X: features
    :param y: target
    :param cv: int
        Number of cross-validation folds. Default is 5.
    :param scoring: list
        Scoring metrics to evaluate the models.
    :param is_classifier: boolean
        True if classification, False if regression.
    :param random_state: int
        Random seed for reproducibility. Default is 40.
    :param n_jobs: int
        Number of CPU cores to use. Default is 2.
    """
    warnings.filterwarnings("ignore", category=FutureWarning)
    warnings.filterwarnings("ignore", category=UserWarning)
    warnings.filterwarnings("ignore", category=ConvergenceWarning)
    warnings.filterwarnings("ignore", category=FitFailedWarning)
    
    print("Base Models....")
    if is_classifier:
        models = {
            "GNB": GaussianNB(),
            "BNB": BernoulliNB(),
            "CATNB": CategoricalNB(),
            "LR": LogisticRegression(),
            "SGD": SGDClassifier(),
            "RID": Ridge(),  # Regression Model - used for continuous variables
            "LAS": Lasso(),  # Regression Model
            "ENET": ElasticNet(),  # Regression Model
            "SVC": SVC(),
            "LSVC": LinearSVC(),
            "KNN": KNeighborsClassifier(),
            "RANEC": RadiusNeighborsClassifier(),
            "CART": DecisionTreeClassifier(),
            "EXTR": ExtraTreeClassifier(),
            "RF": RandomForestClassifier(),
            "BAG": BaggingClassifier(),
            "GBM": GradientBoostingClassifier(),
            "HIST": HistGradientBoostingClassifier(),
            "Adaboost": AdaBoostClassifier(),
            "XGBoost": XGBClassifier(device="cuda"),
            "LightGBM": LGBMClassifier(verbose=-1),
            "Catboost": CatBoostClassifier(verbose=False, task_type="GPU", devices='0:1'),
            "MLP": MLPClassifier(),
        }
    else:
        models = {
            "GPR": GaussianProcessRegressor(),
            "LiR": LinearRegression(),
            "SGD": SGDRegressor(),
            "RID": Ridge(),
            "LAS": Lasso(),
            "ENET": ElasticNet(),
            # "SVR": SVR(),
            #"NUSVR": NuSVR(),
            # "LSVR": LinearSVR(),
            # "KNN": KNeighborsRegressor(),
            "CART": DecisionTreeRegressor(),
            "EXTR": ExtraTreeRegressor(),
            "RF": RandomForestRegressor(),
            "BAG": BaggingRegressor(),
            "GBM": GradientBoostingRegressor(),
            "HIST": HistGradientBoostingRegressor(),
            "Adaboost": AdaBoostRegressor(),
            "XGBoost": XGBRegressor(device="cuda"),
            "LightGBM": LGBMRegressor(verbose=-1),
            # "Catboost": CatBoostRegressor(verbose=False, task_type="GPU", devices='0:1'),
            "MLP": MLPRegressor()
        }

    for name, model in models.items():
        print(f"################## {name} ################## ")
        try:
            for score_param in scoring:
                X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=random_state)
                cv_results = cross_validate(model, X_train, y_train, cv=cv, scoring=score_param, n_jobs=n_jobs)
                print(f"{score_param}: {abs(round(cv_results['test_score'].mean(), 4))} ({name}) ")
        except ValueError as e:
            print("None")
            continue
        print()

```

```python
train = df[df['FloodProbability'].notnull()]
test = df[df['FloodProbability'].isnull()]

X_train = train.drop(["id",'FloodProbability'], axis=1)
y_train = train['FloodProbability']

X_test = test.drop(["id",'FloodProbability'], axis=1)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import backend as K
import optuna as opt

from keras import models, layers, optimizers, callbacks, regularizers

def r2_score(y_true, y_pred):
    ss_res = K.sum(K.square(y_true - y_pred))
    ss_tot = K.sum(K.square(y_true - K.mean(y_true)))
    return (1 - ss_res / (ss_tot + K.epsilon()))


# TensorFlow'un GPU'yu gÃ¶rÃ¼p gÃ¶rmediÄŸini kontrol edin
print("TensorFlow version:", tf.__version__)
print("GPU device available:", tf.config.list_physical_devices('GPU'))
```

```python
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
    except RuntimeError as e:
        print(e)
```

```python

# Not: CPU daha hÄ±zlÄ±, ama HER ZAMAN DEÄžÄ°L. Bu yÃ¼zden GPU kullanÄ±lacak tuning sÄ±rasÄ±nda.

# early_stopping = callbacks.EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
batch_size = 1024

model = models.Sequential()
model.add(layers.Dense(64, input_shape=(X_train.shape[1],), activation='relu'))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(1, activation='linear'))

# Modeli derleme
model.compile(optimizer='adam', loss='mean_squared_error', metrics=[r2_score])
```

```python
# Modeli eÄŸitme
history = model.fit(X_train, y_train, epochs=100, batch_size=batch_size)
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(y=history.history['r2_score'], x=range(len(history.history['r2_score'])), label='train')
# sns.lineplot(y=history.history['val_r2_score'], x=range(len(history.history['val_r2_score'])), label='validation')
plt.show(block=True)
```

```python
pred = model.predict(X_test, batch_size=batch_size)
```
