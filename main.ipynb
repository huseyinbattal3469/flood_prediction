```python
train = df[df['FloodProbability'].notnull()]
test = df[df['FloodProbability'].isnull()]

X_train = train.drop(["id",'FloodProbability'], axis=1)
y_train = train['FloodProbability']

X_test = test.drop(["id",'FloodProbability'], axis=1)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import backend as K
import optuna as opt

from keras import models, layers, optimizers, callbacks, regularizers

def r2_score(y_true, y_pred):
    ss_res = K.sum(K.square(y_true - y_pred))
    ss_tot = K.sum(K.square(y_true - K.mean(y_true)))
    return (1 - ss_res / (ss_tot + K.epsilon()))


# TensorFlow'un GPU'yu gÃ¶rÃ¼p gÃ¶rmediÄini kontrol edin
print("TensorFlow version:", tf.__version__)
print("GPU device available:", tf.config.list_physical_devices('GPU'))
```

```python
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
    except RuntimeError as e:
        print(e)
```

```python

# Not: CPU daha hÄ±zlÄ±, ama HER ZAMAN DEÄÄ°L. Bu yÃ¼zden GPU kullanÄ±lacak tuning sÄ±rasÄ±nda.

# early_stopping = callbacks.EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
batch_size = 1024

model = models.Sequential()
model.add(layers.Dense(64, input_shape=(X_train.shape[1],), activation='relu'))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(1, activation='linear'))

# Modeli derleme
model.compile(optimizer='adam', loss='mean_squared_error', metrics=[r2_score])
```

```python
# Modeli eÄitme
history = model.fit(X_train, y_train, epochs=100, batch_size=batch_size)
```

```python
print(model.output_shape)
print(model.summary())
print(model.get_config())
# print(model.get_weights())

loss, r2 = model.evaluate(X_train, y_train, batch_size=batch_size)
print(f'Test loss: {loss}')
print(f'Test R^2 score: {r2}')
```

```python
pred = model.predict(X_test, batch_size=batch_size)
```

```python
submission_file = pd.DataFrame(data=pred, index=test['id'], columns=["FloodProbability"])

submission_file.to_csv("submission.csv")
```
